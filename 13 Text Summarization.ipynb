# Text Summarization

Text summarization is an NLP task that involves creating a concise and coherent summary of a larger text document while retaining the key information and main ideas. There are two primary types of text summarization:

- **Extractive Summarization:** Selects and extracts important sentences or phrases from the original text to create the summary. This method does not generate new sentences but relies on existing content.

- **Abstractive Summarization:** Generates new sentences that convey the main ideas of the original text. This method uses more complex language models to understand and reformulate the content.

```python
# Install necessary libraries
!pip install nltk gensim

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
import gensim
from gensim.summarization import summarize

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Sample text
text = """
    Text summarization is the process of condensing a piece of text to its essential points 
    while retaining the key information and meaning. It can be broadly categorized into two 
    types: extractive and abstractive summarization. Extractive Summarization involves 
    selecting important sentences, phrases, or sections from the original text and concatenating 
    them to create a summary. It relies on identifying the most significant portions of the text 
    without altering the original wording. Abstractive Summarization generates a summary by 
    understanding the main concepts of the text and then expressing them in a new way, often 
    with different wording and structure. It is more complex and closer to human summarization.
    """

# Tokenize the text into sentences
sentences = sent_tokenize(text)

# Extractive summarization using gensim
summary = summarize(text, word_count=50)

print("Original Text:\n", text)
print("\nExtractive Summary:\n", summary)
```
